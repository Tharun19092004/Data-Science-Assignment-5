# TASK-1: Data Cleaning and Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Cardiotocographic.csv')
df
```

```python
# ●	Handle missing values appropriately (e.g., imputation, deletion).
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
impute=SimpleImputer(missing_values=np.nan,strategy=np.median)
df=pd.DataFrame(impute.fit_transform(df))
df
```

```python
df.isnull().sum()
```

```python
# ●	Identify and correct any inconsistencies in data types (e.g., numerical values stored as strings).
df.info()
```

```python
# ●	Detect and treat outliers if necessary.
```

```python
df.boxplot()
plt.show()
```

```python
def outlier_capping(df,columns):
    Q1=df[columns].quantile(0.25)
    Q3=df[columns].quantile(0.75)
    IQR=Q3-Q1
    Lower_extreme=Q1-1.5*IQR
    Upper_extreme=Q3+1.5*IQR 
    df[columns]=df[columns].apply(lambda x: Lower_extreme if x<Lower_extreme else Upper_extreme if x>Upper_extreme else x)
for col in df.select_dtypes(['int','float']).columns:
    outlier_capping(df,col)
```

```python
df.boxplot()
plt.show()
```

# Bar charts

```python
# Use bar charts or pie charts to display the frequency of categories for categorical variables.
for col in df.columns:
    plt.figure(figsize=(7,5))
    df[col].value_counts().plot(kind='bar')
    plt.show()
```

# correlation heatmaps

```python
plt.figure(figsize=(16,12))
sns.heatmap(df.corr(),annot=True,linecolor='white',linewidths=2,cmap='coolwarm')
plt.show()
```

# Pair plots

```python
sns.pairplot(df,diag_kind='hist')
plt.show()
```
